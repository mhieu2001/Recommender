```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
books = pd.read_csv(r"C:\Users\nguye\OneDrive\Desktop\Recommender system\BX-Books.csv", sep =";",error_bad_lines=False,encoding='latin-1')
users = pd.read_csv(r"C:\Users\nguye\OneDrive\Desktop\Recommender system\BX-Users.csv", sep =";",error_bad_lines=False,encoding='latin-1')
rates = pd.read_csv(r"C:\Users\nguye\OneDrive\Desktop\Recommender system\BX-Book-Ratings.csv", sep =";",error_bad_lines=False,encoding='latin-1')
```

```python
books.rename(columns={"Book-Title":'title',
                      'Book-Author':'author',
                     "Year-Of-Publication":'year',
                     "Publisher":"publisher",
                     "Image-URL-L":"image_url"},inplace=True)

books.head(2)
```

```python
books.shape
```

```python
users.rename(columns={"User-ID":'user_id',
                      'Location':'location',
                     "Age":'age'},inplace=True)
users.head(2)
```

```python
users. shape
```

```python
rates.rename(columns={"User-ID":'user_id',
                      'Book-Rating':'rating'},inplace=True)
rates.head(2)
```

```python
rates.shape
```

```python
x = rates['user_id'].value_counts() > 200
y= x[x].index
```

```python
rates = rates[rates['user_id'].isin(y)]
```

```python
rates.head()
```

```python
rates.shape
```

```python
rates_with_books = rates.merge(books, on='ISBN')
rates_with_books.head()
```

```python
num_rating=rates_with_books.groupby('title')['rating'].count().reset_index()
num_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
num_rating.head()
```

```python
final_rating = rates_with_books.merge(num_rating, on='title')
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values= 'rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

# Training Model

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
type(book_sparse)
```

```python
# Now import our clustering algoritm which is Nearest Neighbors this is an unsupervised ml algo
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
n=int(input())
distance, suggestion = model.kneighbors(book_pivot.iloc[n,:].values.reshape(1,-1), n_neighbors=6 )
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

# Testing Model 

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name=input()
recommend_book(book_name)
```
